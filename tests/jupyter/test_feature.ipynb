```python
import pyopenms as oms
from OpenMSTools.old import align, elution_peak_detection, load_exp_ms_files, mapping_features, mass_traces_detection
```

```python
exps,_ = load_exp_ms_files([
    "../data/raw_files/Metabolomics_1.mzML",
    "../data/raw_files/Metabolomics_2.mzML",
],key_type="index")
```

```python
mass_traces = mass_traces_detection(exps)
```

```python
mass_traces = elution_peak_detection(mass_traces)
```

```python
features,_ = mapping_features(mass_traces)
```

```python
align(exps, features, 0)
```

```python
rebuild_features: dict[int, oms.FeatureMap] = {}
for exp_id, feature_map in features.items():
    if exp_id not in rebuild_features:
        rebuild_features[exp_id] = oms.FeatureMap()
    for feature in feature_map:
        feature: oms.Feature
        rebuild_feature = oms.Feature()
        rebuild_feature.setMZ(feature.getMZ())
        # rebuild_feature.setCharge(feature.getCharge())
        rebuild_feature.setRT(feature.getRT())
        rebuild_feature.setIntensity(feature.getIntensity())
        # rebuild_feature.setOverallQuality(feature.getOverallQuality())
        rebuild_features[exp_id].push_back(
            rebuild_feature
        )
```
